```python
from numpy import random
import matplotlib.pyplot as plt
from scipy.special import factorial
from PIL import Image
PATH = "img/"
```

```python
sample_size = [10, 50, 1000]
loc, scale = 0.0, 1.0
bins_num = 20
```

## Нормальное распределение:

```python
Image.open(fp = PATH + "norm.jpg")
```

```python
i=1
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Normal Distribution', y = 1.05, fontsize = 20)
for size in sample_size:
    plt.subplot(1, 3, i)   
    i += 1
    
    sample = np.random.normal(loc, scale, size)
    n, bins, patches = plt.hist(sample, bins_num, density=1, facecolor='blue', edgecolor='black', alpha=0.3)
    fun_n = 1/(scale * np.sqrt(2 * np.pi)) * np.exp( - (bins - loc)**2 / (2 * scale**2) )
   
    plt.plot(bins, fun_n, color='b', linewidth=1)
    plt.subplots_adjust(wspace=0.5)
    plt.title('$\mu=0$, $\sigma=1$, N=%i' %size)
    plt.xlabel('numbers')
    plt.ylabel('density')
    
    
```

## Распределение Коши:

```python
Image.open(fp = PATH + "Cauchy.jpg")
```

```python
i=1
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Cauchy Distribution', y = 1.05, fontsize = 20)

for size in sample_size:
    plt.subplot(1, 3, i) 
    i += 1
    
    sample = np.random.standard_cauchy(size)
    n, bins, patches = plt.hist(sample, bins_num, density=1, facecolor='blue', edgecolor='black', alpha=0.3)
    fun_c = 1 / (np.pi * (bins * bins + 1))
    
    plt.plot(bins, fun_c, color='b', linewidth=1)
    plt.subplots_adjust(wspace=0.5)
    plt.title('N=%i' %size)
    plt.xlabel('numbers')
    plt.ylabel('density')
    
    
```

## Распределение Лапласа:

```python
Image.open(fp = PATH + "Laplace.jpg")
```

```python
i=1
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Laplace Distribution', y = 1.05, fontsize = 20)

for size in sample_size:
    plt.subplot(1, 3, i) 
    i += 1
    
    sample = np.random.laplace(loc, 1 / np.sqrt(2), size)
    n, bins, patches = plt.hist(sample, bins_num, density=1, facecolor='blue', edgecolor='black', alpha=0.3)
    fun_l = 1 / np.sqrt(2) * np.exp(-np.sqrt(2) * np.fabs(bins))
    
    plt.plot(bins, fun_l, color='b', linewidth=1)
    plt.subplots_adjust(wspace=0.5)
    plt.title('N=%i' %size)
    plt.xlabel('numbers')
    plt.ylabel('density')
    
    
```

## Распределение Пуассона:

```python
Image.open(fp = PATH + "Poisson.jpg")
```

```python
i=1
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Poisson Distribution', y = 1.05, fontsize = 20)

for size in sample_size:
    plt.subplot(1, 3, i) 
    i += 1
    
    sample = np.random.poisson(10, size)
    n, bins, patches = plt.hist(sample, bins_num, density=1, facecolor='blue', edgecolor='black', alpha=0.3)
    fun_p = np.power(10, bins) * np.exp(-10) / factorial(bins)
    
    plt.plot(bins, fun_p, color='b', linewidth=1)
    plt.subplots_adjust(wspace=0.5)
    plt.title('N=%i' %size)
    plt.xlabel('numbers')
    plt.ylabel('density')
    
    
```

## Равномерное распределение:

```python
Image.open(fp = PATH + "Uniform.jpg")
```

```python
i=1
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Uniform Distribution', y = 1.05, fontsize = 20)

for size in sample_size:
    plt.subplot(1, 3, i)  
    i += 1
    
    sample = np.random.uniform(-np.sqrt(3), np.sqrt(3), size)
    n, bins, patches = plt.hist(sample, bins_num, density=1, facecolor='blue', edgecolor='black', alpha=0.3)
    pdf = []
    ar = np.arange(-2., 2., 0.01)
    for elem in ar:
        pdf.append(1 / (2 * np.sqrt(3))) if np.fabs(elem) <= np.sqrt(3) else pdf.append(0)
    
    plt.plot(ar, pdf, color='b', linewidth=1)
    plt.subplots_adjust(wspace=0.5)
    plt.title('N=%i' %size)
    plt.xlabel('numbers')
    plt.ylabel('density')
    
    
```
